```python
import pandas
import numpy

analiseDrinks = pandas.read_csv("C:\GIT\DataScience\Datasets\drinks.csv")
analiseDrinks.head(5)
```

```python
#Categorizando consumo de cerveja, por uma funcao

def consumo (litros):
    if litros <= 75:
        return 'low'
    elif litros >= 151:
        return 'high'
    else:
        return 'middle'
```

```python
#Dicionar uma coluna consumo no dataFrame para receber o resultado
                         #chamando a funcao para executar no dataFrame
analiseDrinks['consumo'] = analiseDrinks['beer_servings'].apply(consumo)
```

```python
#FAZENDO AS MESMAS OPERAÇÕES UTILIZANDO NUMPY - aggregate

consumoNumpy = analiseDrinks.groupby(['consumo', 'continent'])
consumoNumpy.aggregate(numpy.sum)
```

```python
#agrupando por continent
#refazendo os passo acima mas agrupando por continent em primeiro ao inves de consumo

#transormando colunas em index - menos country
consumoContinentNumpy = analiseDrinks.set_index(['continent', 'consumo','beer_servings','spirit_servings',
                                                'wine_servings', 'total_litres_of_pure_alcohol'])

consumoContinentNumpy = consumoContinentNumpy.groupby(level=['continent', 'consumo']).sum()
consumoContinentNumpy.head(20) #trouxe todos os paises classficados por consumo e categorizado por continent
```
