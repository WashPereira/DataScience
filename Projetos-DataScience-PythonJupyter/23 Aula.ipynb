```python
import matplotlib.pyplot as matplot
%matplotlib inline
```

```python
x = []
y1 = []
y2 = []
y3 = []

for i in range (-200, 200, 30):
    x.append(i)
    y1.append(i)
    y2.append(i * 2)
    y3.append(i + 50)
```

```python
# ADCICONADO DESCRIÇÃO AOS EIXOS

matplot.xlabel('Eixo x', color='red', fontsize=12, horizontalalignment='right')
matplot.ylabel('Eixo y', color='blue', fontsize=12, horizontalalignment='left')
matplot.title('Titulo', color='black', fontsize=15, horizontalalignment='center')

matplot.plot(x,y3, c='black', ls =':', lw = 4, ms=20, fillstyle='left') #ls = line style
matplot.plot(x,y2, c='black', ls ='--', lw = 4, ms=20, fillstyle='left') #ls = line style
```

```python
# TEXTO MATEMATICO

matplot.xlabel('Eixo x', color='red', fontsize=12, horizontalalignment='right')
matplot.ylabel('Eixo y', color='blue', fontsize=12, horizontalalignment='left')
matplot.title(r'Titulo $\frac{5}{4}$', color='black', fontsize=15, horizontalalignment='center')

matplot.plot(x,y3, c='black', ls =':', lw = 4, ms=20, fillstyle='left') #ls = line style
matplot.plot(x,y2, c='black', ls ='--', lw = 4, ms=20, fillstyle='left') #ls = line style
```

```python
# LEGENDA e labels

#titilo
matplot.title(r'Titulo $\frac{5}{4}$', color='black', fontsize=15, horizontalalignment='center')

#grafico
matplot.plot(x,y3, c='black', ls =':', lw = 4, ms=20, fillstyle='left',  label='primeiro') #label
matplot.plot(x,y2, c='black', ls ='--', lw = 4, ms=20, fillstyle='left', label='segundo') #label

#legenda
matplot.legend()

#posição da legenda
matplot.legend(loc='lower right')

#legenda de eixos
matplot.xlabel('Eixo x', color='red', fontsize=12, horizontalalignment='right')
matplot.ylabel('Eixo y', color='blue', fontsize=12, horizontalalignment='left')


```

```python
#TICKERS DE X E Y

#tickers
eixoX = ['Janeiro', 'Abril', 'Julho']
eixoY = ['fraco', 'regular', 'neutro', 'bom', 'exelente'] 
matplot.xticks([-200,0,200], eixoX, rotation = 30)#variação do eixo x no grafico (numero min, medio, max)
matplot.yticks([-400,-100,0,100, 400], eixoY, rotation =30)#           Y no grafico (numero min, medio, max)

#titilo
matplot.title(r'Titulo $\frac{5}{4}$', color='black', fontsize=15, horizontalalignment='center')

#grafico
matplot.plot(x,y3, c='black', ls =':', lw = 1, ms=50, fillstyle='left',  label='primeiro') #label
matplot.plot(x,y2, c='black', ls ='--', lw = 1, ms=50, fillstyle='left', label='segundo') #label

#legenda
matplot.legend()

#posição da legenda
matplot.legend(loc='lower right')

#legenda de eixos
matplot.xlabel('Eixo x', color='red', fontsize=12, horizontalalignment='right')
matplot.ylabel('Eixo y', color='blue', fontsize=12, horizontalalignment='left')
```
