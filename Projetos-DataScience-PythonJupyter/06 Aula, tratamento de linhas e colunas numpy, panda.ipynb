```python
#as colunas tambem podem ser renomadas no momento da abertura/leitura do aquivo em forma de lista

arquivoFilmes2 = panda.read_csv('C:\GIT\DataScience\Datasets\movies.csv', names = ['star_rating', 'title',
                                                                                     'content_rating', 'genre', 
                                                                                     'durating', 'actors_list'])
arquivoFilmes2.head()
```

```python
#REMOVENDO LINHAS DA TABELA

novoArquivoFilmes2 = novoArquivoFilmes.drop([0, 1], axis=0) #removeno linhas 0 e 1
novoArquivoFilmes2
```
